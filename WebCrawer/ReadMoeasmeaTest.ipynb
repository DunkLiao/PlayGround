```python
import requests
import warnings
import json
import pprint
warnings.filterwarnings('ignore')
url = "https://sme.moeasmea.gov.tw/startup/modules/highlight/getdata1.php"

payload = "pageIndex=1"
headers = {
    'User-Agent': "Mozilla/5.0 (Windows NT 6.1; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/81.0.4044.92 Safari/537.36",
    'Referer': "https://sme.moeasmea.gov.tw/startup/modules/highlight/index.php"
    }

response = requests.request("POST", url, data=payload, headers=headers,verify = False)
response.encoding = 'utf-8'
data = json.loads(response.text)
#顯示讀取資料
pprint.pprint(data)
```

```python
# 1. 這個 API 一次會回傳幾筆資料？每一筆資料包含哪些欄位？
print('筆數:%d'% int(len(data)))
print('包含欄位:%s' % data.keys())
```
