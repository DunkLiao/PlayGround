```python
import requests
import pandas as pd
import json
import warnings
warnings.filterwarnings('ignore')
```

```python
consdb_url = "https://www.taiwanstat.com/waters/latest"
consdb_content = requests.get("{0}".format(consdb_url),verify=False)
data_json = json.loads(consdb_content.content)
```

```python
key_item = ""
for key in data_json[0].keys():
    key_item = key
    break
df_enterprise  = pd.DataFrame(columns= data_json[0][key_item].keys())    

for item in data_json:
    for keyitem in item.keys():
        key = item[keyitem]
        df_enterprise = df_enterprise.append(pd.Series([key['baseAvailable'],key['id'],key['updateAt'],key['volumn'],\
                                                       key['percentage'],key['daliyOverflow'],key['daliyInflow'],\
                                                        key['daliyNetflow'],key['name']],\
                                                       index=df_enterprise.columns ), ignore_index=True)
df_enterprise.sort_values(by=['percentage'])        
```
