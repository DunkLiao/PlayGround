#### 呼叫所需套件

```python
import requests
import pandas as pd
import json
import warnings
warnings.filterwarnings('ignore')
```

```python
# 指定網址:社會創新企業資料庫
consdb_url = "https://si.taiwan.gov.tw/api/orgs?page=1&limit=12"
consdb_content = requests.get("{0}".format(consdb_url),verify=False)
data_json = json.loads(consdb_content.content)
df_enterprise  = pd.DataFrame(columns=['id','title','url','description'])
total_page = int(data_json['meta']['total_page'])

for i in range(total_page):
    query_url = "https://si.taiwan.gov.tw/api/orgs?page={0}&limit=12".format(str(i+1))
    consdb_content = requests.get("{0}".format(query_url),verify=False)
    data_json = json.loads(consdb_content.content)
    for key in data_json['data']:    
        df_enterprise = df_enterprise.append(pd.Series([key['id'],key['title'],key['url'],key['description']],\
                                                       index=df_enterprise.columns ), ignore_index=True)

#顯示彙整結果
df_enterprise
```
