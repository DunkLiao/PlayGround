```python
# #############################################################################
# 本題參數設定，請勿更改
seed = 0    # 亂數種子數
# #############################################################################
# TODO

import numpy as np
from sklearn.datasets import load_digits

# 載入手寫數字資料集
digits = load_digits()
X_digits = digits.data   #TODO    數值特徵
y_digits = digits.target   #TODO    數字類別

# 特徵標準化(scale/StandardScaler)
from sklearn.preprocessing import scale
from sklearn.preprocessing import StandardScaler
# TODO
scalar = StandardScaler()

data = scale(X_digits)

# 取出資料集的數字類別數
# n_digits = 10
n_digits = len(np.unique(digits.target))
print(n_digits)

# 建立兩個 K-Means 模型，除以下參數設定外，其餘為預設值
# #############################################################################
# kmean1: init='k-means++', n_clusters=n_digits, n_init=10, random_state=seed
# kmean2: init='random', n_clusters=n_digits, n_init=10, random_state=seed
# #############################################################################
from sklearn.cluster import KMeans
# TODO
kmean1 = KMeans(init='k-means++', n_clusters=n_digits, n_init=10, random_state=seed)
kmean2 = KMeans(init='random', n_clusters=n_digits, n_init=10, random_state=seed)
kmean1.fit(data)
kmean2.fit(data)
# 利用 PCA 結果建立 K-Means 模型，除以下參數設定外，其餘為預設值
# #############################################################################
# pca: n_components=n_digits, random_state=seed
# kmean3: init=pca.components_, n_clusters=n_digits, n_init=1, random_state=seed
# #############################################################################
from sklearn.decomposition import PCA
# TODO
pca = PCA(n_components=n_digits, random_state=seed)
pca.fit(data)
kmean3 = KMeans(init=pca.components_, n_clusters=n_digits, n_init=1, random_state=seed)
kmean3.fit(data)
# 分別計算上述三個 K-Means 模型的輪廓係數(Silhouette coefficient)與
# 分類準確率(accuracy)，除以下參數設定外，其餘為預設值
# #############################################################################
# silhouette_score: metric='euclidean'
# #############################################################################
from sklearn.metrics import silhouette_score
from sklearn.metrics import accuracy_score
si_score = []
ac_score = []
lst_name = ['K-Mean (k-means++)', 'K-Means (random)', 'K-Means (PCA-based)']
# TODO
print(lst_name[0])
si_score.append(round(silhouette_score(data,kmean1.labels_),4))
ac_score.append(round(accuracy_score(y_digits,kmean1.labels_),4))
print(lst_name[1])
si_score.append(round(silhouette_score(data,kmean2.labels_),4))
ac_score.append(round(accuracy_score(y_digits,kmean2.labels_),4))
print(lst_name[2])
si_score.append(round(silhouette_score(data,kmean3.labels_),4))
ac_score.append(round(accuracy_score(y_digits,kmean3.labels_),4))

# 進行 PCA 降維後再做 K-Means，除以下參數設定外，其餘為預設值
# #############################################################################
# kmeans: init='k-means++', n_clusters=n_digits, n_init=10, random_state=seed
# PCA: n_components=2, random_state=seed
# #############################################################################
# TODO
pca = PCA(n_components=2, random_state=seed)
reduced_data = pca.fit_transform(data)
kmeans = KMeans(init='k-means++', n_clusters=n_digits, n_init=10, random_state=seed)
kmeans.fit(reduced_data)
print('PCA+KMeans Silhouette=   ',round(silhouette_score(data,kmeans.labels_),4)      )
print('Accuracy=    ',round(accuracy_score(y_digits,kmeans.labels_),4)                 )
si_score.append(round(silhouette_score(data,kmeans.labels_),4))
ac_score.append(round(accuracy_score(y_digits,kmeans.labels_),4))
print('模型A的輪廓係數（四捨五入取至小數點後第四位）:',si_score[0])
print('模型B的輪廓係數（四捨五入取至小數點後第四位）:',si_score[1])
print('模型C的輪廓係數（四捨五入取至小數點後第四位）:',si_score[2])
print('四個模型的最大準確率（四捨五入取至小數點後第四位）:',max(ac_score))
```
