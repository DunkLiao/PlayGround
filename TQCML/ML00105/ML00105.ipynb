```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn.feature_selection import f_regression
# 注意引用
from sklearn import preprocessing

NBApoints_data= pd.read_csv("NBApoints.csv")
#TODO


label_encoder_conver = preprocessing.LabelEncoder()
# 注意寫法
Pos_encoder_value = label_encoder_conver.fit_transform(NBApoints_data['Pos'])
print(Pos_encoder_value)
print("\n")

label_encoder_conver = preprocessing.LabelEncoder()
# 注意寫法
Tm_encoder_value = label_encoder_conver.fit_transform(NBApoints_data['Tm'])
print(Tm_encoder_value)

# 注意寫法
train_X = pd.DataFrame({'Pos':Pos_encoder_value,'Age':NBApoints_data['Age'],\
                        'Tm':Tm_encoder_value})
y = NBApoints_data["3P"]                        
NBApoints_linear_model = LinearRegression()
NBApoints_linear_model.fit(train_X, y)
y_pred = NBApoints_linear_model.predict(train_X)


NBApoints_linear_model_predict_result=NBApoints_linear_model.predict(train_X)
print("三分球得球數=",NBApoints_linear_model_predict_result)

y_pred_new = NBApoints_linear_model.predict([[5,28,10]])
# 注意轉型
print("三分球預測得球數=","{:.4f}".format(round(float(y_pred_new),4)))

# 注意引用
from sklearn.metrics import accuracy_score,r2_score
r_squared = r2_score(y,y_pred)
print("R_squared值=","{:.4f}".format(round(r_squared,4)))

# 注意要train與結果比對
(F,p_value) = f_regression(train_X,y)
print("f_regresstion\n")
print("P值=",p_value)
print("\n")
print("Pos的 P-value （P值）是否小於 0.05（信心水準 95%）",p_value[0]<0.05)
print("Age的 P-value （P值）是否小於 0.05（信心水準 95%）",p_value[1]<0.05)
```
