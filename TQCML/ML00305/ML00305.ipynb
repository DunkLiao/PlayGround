```python
# #############################################################################
# 本題參數設定，請勿更改
seed = 0   # 亂數種子數  
# #############################################################################
import warnings
warnings.filterwarnings('ignore')

import numpy as np
import pandas as pd 

# TODO

# 讀取台北市房價資料集
# TODO
input_file = "Taipei_house.csv"
df = pd.read_csv(input_file)
# 注意先複製一份出來
df_source = df.copy(deep=False)

# 對"行政區"進行 one-hot encoding
# TODO
# 注意寫法
df_source = pd.get_dummies(df_source,columns=['行政區'])

# 處理"車位類別"
# TODO
# 注意自訂函式 
def get_car_type(value):
    if value == "無":
        return 0
    else:
        return 1
    
df_source['車位類別'] = df_source['車位類別'].apply(get_car_type)

# 計算 Adjusted R-squared
def adj_R2(r2, n, k):
    """ 函式描述：計算 Adjusted R-squared
    參數：
        r2:R-squared 數值
        n: 樣本數
        k: 特徵數

    回傳：
        Adjusted R-squared
    """
    return r2-(k-1)/(n-k)*(1-r2)

# TODO
# 注意自訂函式    
def get_round_adj_R2(r2, n, k):    
    return round(adj_R2(r2, n, k),4)

def get_int_value(value):
    return round(value,0)

# # 切分訓練集(80%)、測試集(20%)
features= ['土地面積', '建物總面積', '屋齡', '樓層', '總樓層', '用途', 
            '房數', '廳數', '衛數', '電梯', '車位類別', 
            '行政區_信義區', '行政區_大安區', '行政區_文山區','行政區_松山區']
target = '總價'  
# # TODO
# 注意匯入
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error,r2_score
from math import sqrt

X = df_source[features]
y = df_source[target]
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2,\
                                                 random_state=seed)

# 注意設定DataFram為evaluation    
evaluation = pd.DataFrame(columns=[ 'type','adj. R2 (train)',\
                                   'adj. R2 (test)','RMSE (train)','RMSE (test)'])

# 注意取樣本數及特徵數    
sample_train = X_train.shape[0]
sample_test = X_test.shape[0]
feature_len = len(features)
# 複迴歸(參數皆為預設值)
# #########################################################################
# '行政區_信義區', '行政區_大安區', '行政區_文山區','行政區_松山區' 四個特徵是經過
# one-hot encoding 後產生，若欄位名稱不同可自行修改之。
# #########################################################################
from sklearn import linear_model
# TODO
model = linear_model.LinearRegression()
model.fit(X_train,y_train)
y_pred_train = model.predict(X_train)
y_pred_test = model.predict(X_test)
sample_len = X_train

result = {'type':"lm",\
          'adj. R2 (train)':get_round_adj_R2(r2_score(y_train,y_pred_train),\
                                             sample_train,feature_len),\
          'adj. R2 (test)':get_round_adj_R2(r2_score(y_test,y_pred_test),\
                                            sample_test,feature_len),\
          'RMSE (train)':get_int_value(sqrt(mean_squared_error(y_train,y_pred_train))),\
          'RMSE (test)':get_int_value(sqrt(mean_squared_error(y_test,y_pred_test)))}
evaluation = evaluation.append(result,ignore_index=True)
# 脊迴歸(Ridge regression)，除以下參數設定外，其餘為預設值
# #########################################################################
# alpha=10
# #########################################################################
# TODO
# 注意寫法
model = linear_model.Ridge(alpha=10)
model.fit(X_train,y_train)
y_pred_train = model.predict(X_train)
y_pred_test = model.predict(X_test)
sample_len = X_train

result = {'type':"ridge",\
          'adj. R2 (train)':get_round_adj_R2(r2_score(y_train,y_pred_train),\
                                             sample_train,feature_len),\
          'adj. R2 (test)':get_round_adj_R2(r2_score(y_test,y_pred_test),\
                                            sample_test,feature_len),\
          'RMSE (train)':get_int_value(sqrt(mean_squared_error(y_train,y_pred_train))),\
          'RMSE (test)':get_int_value(sqrt(mean_squared_error(y_test,y_pred_test)))}
evaluation = evaluation.append(result,ignore_index=True)

# 多項式迴歸，除以下參數設定外，其餘為預設值
# #########################################################################
# degree=2
# #########################################################################
from sklearn.preprocessing import PolynomialFeatures
# TODO
# 注意寫法
poly = PolynomialFeatures(degree=2)
X_train_poly = poly.fit_transform(X_train)
X_test_poly = poly.fit_transform(X_test)
# 注意寫法
model = linear_model.LinearRegression()
model.fit(X_train_poly,y_train)
y_pred_train = model.predict(X_train_poly)
y_pred_test = model.predict(X_test_poly)
result = {'type':"poly",\
          'adj. R2 (train)':get_round_adj_R2(r2_score(y_train,y_pred_train),\
                                             X_train_poly.shape[0],X_train_poly.shape[1]),\
          'adj. R2 (test)':get_round_adj_R2(r2_score(y_test,y_pred_test),\
                                            X_test_poly.shape[0],X_test_poly.shape[1]),\
          'RMSE (train)':get_int_value(sqrt(mean_squared_error(y_train,y_pred_train))),\
          'RMSE (test)':get_int_value(sqrt(mean_squared_error(y_test,y_pred_test)))}
evaluation = evaluation.append(result,ignore_index=True)

# 多項式迴歸 + Lasso迴歸，除以下參數設定外，其餘為預設值
# #########################################################################
# alpha=10
# #########################################################################
# TODO
# 注意寫法
poly = PolynomialFeatures()
X_train_poly = poly.fit_transform(X_train)
X_test_poly = poly.fit_transform(X_test)
# 注意寫法
model = linear_model.Lasso(alpha=10)
model.fit(X_train_poly,y_train)
y_pred_train = model.predict(X_train_poly)
y_pred_test = model.predict(X_test_poly)
result = {'type':"poly+lasso",\
          'adj. R2 (train)':get_round_adj_R2(r2_score(y_train,y_pred_train),\
                                             X_train_poly.shape[0],X_train_poly.shape[1]),\
          'adj. R2 (test)':get_round_adj_R2(r2_score(y_test,y_pred_test),\
                                            X_test_poly.shape[0],X_test_poly.shape[1]),\
          'RMSE (train)':get_int_value(sqrt(mean_squared_error(y_train,y_pred_train))),\
          'RMSE (test)':get_int_value(sqrt(mean_squared_error(y_test,y_pred_test)))}
evaluation = evaluation.append(result,ignore_index=True)


print(evaluation)
print('對訓練集的最大 Adjusted R-squared: %.4f' % max(evaluation['adj. R2 (train)']))
print('對測試集的最小 RMSE:%d' % min(evaluation['RMSE (test)']))
print('兩個模型對測試集的最大 Adjusted R-squared: %.4f' % 
      max(evaluation.loc[:1, 'adj. R2 (test)']))

''' 預測 '''
# 利用所有資料重新擬合模型，並進行預測
# TODO


#features= ['土地面積', '建物總面積', '屋齡', '樓層', '總樓層', '用途', 
#           '房數', '廳數', '衛數', '電梯', '車位類別', 
#           '行政區_信義區', '行政區_大安區', '行政區_文山區','行政區_松山區']

# TODO
print('對測試集有最大Adjusted R2的模型')
print(evaluation[evaluation['adj. R2 (test)']==max(evaluation['adj. R2 (test)'])])
features_test= [[36, 99, 32, 4, 4, 0, 
          3, 2, 1, 0, 0, 
          0, 0, 0,1]]

poly = PolynomialFeatures()
X_train_poly = poly.fit_transform(X)
X_test_poly = poly.fit_transform(features_test)
model = linear_model.Lasso(alpha=10)
model.fit(X_train_poly,y)
y_pred_test = model.predict(X_test_poly)

print("使用四個模型中對測試集有最大Adjusted R2的模型，再重新擬合所有資料後，\
        預測其房價為何（計算至整數）",\
      get_int_value(y_pred_test[0]))
```
