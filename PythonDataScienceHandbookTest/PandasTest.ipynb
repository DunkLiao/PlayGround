```python
'''資料清洗(Data Munging)'''
import pandas as pd
import numpy as np
import pprint
#顯示版本號
pprint.pprint(pd.__version__)
```

```python
'''序列'''
data = pd.Series([0.25,0.5,0.75,1.0])
pprint.pprint(data)
pprint.pprint(data.values)
pprint.pprint(data.index)
pprint.pprint(data[1])
pprint.pprint(data[1:3])
```

```python
'''通用的數組，可以隨意設定索引'''
data = pd.Series([0.25,0.5,0.75,1.0],index=['a','b','c','d'])
pprint.pprint(data)
data = pd.Series([0.25,0.5,0.75,1.0],index=[33,44,121,45])
pprint.pprint(data)
#以字典設定
population_dict = {'a':123,'b':'afb','c':1.0,'d':'KKK'}
data = pd.Series(population_dict)
pprint.pprint(data)
pprint.pprint(data['a'])
pprint.pprint(data['b':'d'])
pprint.pprint(data[::-1])
#篩選過濾後結果
data_filter = pd.Series(population_dict,index=['a','c'])
pprint.pprint(data_filter)
```

```python
'''DataFrame為特殊的字典'''
area_dict={'Taipei':213,'Losangle':5322,'Taichung':'ppp'}
area = pd.Series(area_dict)
population_dict = {'a':123,'b':'afb','c':1.0,'d':'KKK','Taipei':'zzz'}
population = pd.Series(population_dict)
states = pd.DataFrame({'population':population,'area':area})
pprint.pprint(states)
#索引標籤
pprint.pprint(states.index)
#欄位名稱
pprint.pprint(states.columns)
pprint.pprint(states['area'])
#建立DataFrame對象
pprint.pprint(pd.DataFrame(population,columns=['population']))
pprint.pprint(pd.DataFrame([{'a':1,'b':3},{'b':8,'c':9}]))
#設定行列
pprint.pprint(pd.DataFrame(np.random.rand(3,2),columns=['foo','bar'],index=['a','b','c']))
```

```python
'''索引，不可變更且為有序集合'''
ind = pd.Index([2,3,4,11,89])
pprint.pprint(ind)
pprint.pprint(ind[3:])
pprint.pprint(ind[:2])
pprint.pprint(ind[::2])
#邏輯操作
ind_a = pd.Index([2,3,4,11,89])
ind_b = pd.Index([2,3,5,11,87])
#交集
pprint.pprint(ind_a & ind_b)
#聯集
pprint.pprint(ind_a | ind_b)
#差集
pprint.pprint(ind_a ^ ind_b)
```

```python
'''數值取值與選擇'''
data = pd.Series([0.25,0.5,0.75,1.0],index=['a','b','c','d'])
pprint.pprint(data['a'])
pprint.pprint('a' in data)
pprint.pprint(data.keys())
pprint.pprint(list(data.items()))
#新增字典項目
data['e']=1.25
pprint.pprint(data)
#切片
pprint.pprint(data['a':'c'])
#隱式索引
pprint.pprint(data[0:2])
#掩碼
pprint.pprint(data[(data >0.3) & (data <0.8)])
#花哨
pprint.pprint(data[['a','e']])
```

```python
'''索引器'''
data = pd.Series(['a','b','c'],index=[1,3,5])
pprint.pprint(data)
#顯式
pprint.pprint(data[1])
#隱式，容易與現行索引混淆
pprint.pprint(data[1:3])
#設定索引器(顯式)
pprint.pprint(data.loc[1])
pprint.pprint(data.loc[1:3])
#設定索引器(隱式)
pprint.pprint(data.iloc[1])
pprint.pprint(data.iloc[1:3])
```

```python
'''DataFrame資料選擇'''
area = pd.Series({'a':1,'b':2,'c':3})
pop = pd.Series({'a':'Jack','b':'May','c':'Alice'})
data = pd.DataFrame({'area':area,'pop':pop})
pprint.pprint(data)
#選取資料
pprint.pprint(data['area'])
pprint.pprint(data.area)
pprint.pprint(data.area is data['area'])
#已有pop方法，所以不相同
pprint.pprint(data.pop is data['pop'])
#增加列
data['desc'] = data['area']/10
pprint.pprint(data)
pprint.pprint(data.values)
pprint.pprint(data.T)
pprint.pprint(data.values[0])
pprint.pprint(data['b':'c'])
pprint.pprint(data[1:2])
pprint.pprint(data[data.area>2])
```

```python
'''缺失值'''
#物件缺失(None)
vals1 = np.array([1,None,3,4])
pprint.pprint(vals1)
#數值缺失(NaN)，特殊浮點數
vals2 = np.array([1,np.nan,3,4])
pprint.pprint(vals2)
pprint.pprint(vals2.dtype)
pprint.pprint(vals2.sum())
#忽略NaN
pprint.pprint(np.nansum(vals2))
pprint.pprint(np.nanmax(vals2))
pprint.pprint(np.nanmin(vals2))
pprint.pprint(np.nanmean(vals2))
pprint.pprint(np.nanmedian(vals2))
#在Pandas中，Nan及None可以等價交換
pprint.pprint(pd.Series([1,np.nan,2,None]))
```

```python
'''處理缺失值'''
data = pd.Series([1,np.nan,2,None])
#建立判斷缺失值的掩碼
pprint.pprint(data.isnull())
#取出缺失值部分
pprint.pprint(data[data.isnull()])
#取出非缺失值部分
pprint.pprint(data[data.notnull()])
#去除缺失值
data2 = data.dropna()
pprint.pprint(data2)
#在DataFrame要注意軸
df = pd.DataFrame(np.arange(9).reshape((3,3)))
df.iloc[0,1]= np.nan
df.iloc[2,0]= np.nan
pprint.pprint(df)
pprint.pprint('!!!!!!!!!!!!!!!!!')
pprint.pprint(df.dropna())
pprint.pprint('!!!!!!!!!!!!!!!!!')
pprint.pprint(df.dropna(axis='columns'))
pprint.pprint('!!!!!!!!!!!!!!!!!')
pprint.pprint(df.dropna(axis='rows'))
pprint.pprint('!!!!!!!!!!!!!!!!!')
#要整列都是缺失值
pprint.pprint(df.dropna(axis='rows',how='all'))
pprint.pprint('!!!!!!!!!!!!!!!!!')
#非缺失值最小數量，大於刪除
pprint.pprint(df.dropna(axis='rows',thresh=2))
pprint.pprint('!!!!!!!!!!!!!!!!!')
pprint.pprint(df.dropna(axis='rows',thresh=3))

#填補缺失值
pprint.pprint(data.fillna(0))
#以前一個值補滿
pprint.pprint(data.fillna(method='ffill'))
#以後一個值補滿
pprint.pprint(data.fillna(method='bfill'))
#指定軸
pprint.pprint(df.fillna(method='ffill',axis=1))
pprint.pprint(df.fillna(method='bfill',axis=0))
```

```python
'''合併數據'''
def make_df(cols,ind):
    data = {c:[str(c)+str(i) for i in ind] for c in cols}
    return pd.DataFrame(data,ind)
pprint.pprint(make_df('ABC',range(3)))
ser1 = pd.Series(['a','b','c'],index=[1,2,3])
ser2 = pd.Series(['d','e','f'],index=[4,5,6])
pprint.pprint(pd.concat([ser1,ser2]))
#合併高維度(預設axis=0，row)
df1 = make_df('AB',[1,2])
df2 = make_df('AB',[3,4])
pprint.pprint(df1)
pprint.pprint(df2)
pprint.pprint(pd.concat([df1,df2]))
df3 = make_df('AB',[0,1])
df4 = make_df('CD',[0,1])
pprint.pprint(df3)
pprint.pprint(df4)
pprint.pprint(pd.concat([df3,df4],axis=1))
#合併索引重複
x = make_df('AB',[0,1])
y = make_df('AB',[2,3])
y.index = x.index
pprint.pprint(x)
pprint.pprint(y)
pprint.pprint(pd.concat([x,y]))
#第一種:抓取重複錯誤
try:
    pd.concat([x,y],verify_integrity=True)
except ValueError as e:
    pprint.pprint("Value Error:{0}".format(e))
#第二種:合併後產生新索引
pprint.pprint(pd.concat([x,y],ignore_index=True))
#第三種:設定多重索引
pprint.pprint(pd.concat([x,y],keys=['x','y']))
```

```python
'''類似join的合併'''
df5 = make_df('ABC',[1,2])
df6 = make_df('BCD',[3,4])
pprint.pprint(df5)
pprint.pprint(df6)
pprint.pprint(pd.concat([df5,df6]))
#join
pprint.pprint(pd.concat([df5,df6],join='inner'))
#設定合併欄位
pprint.pprint(pd.concat([df5,df6],join_axes=[df5.columns]))
#append(會建立新物件)
pprint.pprint(df5.append(df6))
```

```python
'''merge'''
df1 = pd.DataFrame({'employee':['Bob','Jack','Lisa','Sue'],\
                    'group':['Accounting','Engineering','Engineering','HR']})
df2 = pd.DataFrame({'employee':['Lisa','Bob','Jack','Sue'],\
                    'hire_date':['2004','2008','2012','2014']})
pprint.pprint(df1)
pprint.pprint(df2)
#合併成同一個(1對1)
df3 = pd.merge(df1,df2)
pprint.pprint(df3)
#1對多
df4 = pd.DataFrame({'group':['Accounting','Engineering','HR'],\
                    'supervisor':['Carly','Guido','Steve']})
pprint.pprint(df4)
pprint.pprint(pd.merge(df3,df4))
#多對多
df5 = pd.DataFrame({'group':['Accounting','Accounting','Engineering','Engineering','HR','HR'],\
                    'skills':['math','spreadsheets','coding','linux','spreadsheets','organization']})
pprint.pprint(df5)
pprint.pprint(pd.merge(df1,df5))

#設定合併的鍵值(key)，只能在有相同欄位名稱時使用
pprint.pprint(pd.merge(df1,df2,on='employee'))

#設定左與右Dataframe不同欄位名稱做merge
df3 = pd.DataFrame({'name':['Bob','Jack','Lisa','Sue'],\
                    'salary':[70000,80000,120000,90000]})
pprint.pprint(df1)
pprint.pprint(df3)
pprint.pprint(pd.merge(df1,df3,left_on='employee',right_on='name'))
#將多餘的欄位移除
pprint.pprint(pd.merge(df1,df3,left_on='employee',right_on='name').drop('name',axis=1))
#設定合併索引(left及right)
df1a = df1.set_index('employee')
df2a = df2.set_index('employee')
pprint.pprint(df1a)
pprint.pprint(df2a)
pprint.pprint(pd.merge(df1a,df2a,left_index=True,right_index=True))
#跟上一句同等效力
pprint.pprint(df1a.join(df2a))
#混合使用
pprint.pprint(pd.merge(df1a,df3,left_index=True,right_on='name'))
```

```python
'''集合操作規則'''
df6 = pd.DataFrame({'name':['Peter','Paul','Mary'],\
                    'food':['fish','beans','bread']},\
                   columns=['name','food'])
df7 = pd.DataFrame({'name':['Mary','Joseph'],\
                    'drink':['wine','bear']},\
                   columns=['name','drink'])
pprint.pprint(df6)
pprint.pprint(df7)
pprint.pprint(pd.merge(df6,df7))
#交集
pprint.pprint("--------------------")
pprint.pprint(pd.merge(df6,df7,how='inner'))
#外積
pprint.pprint("--------------------")
pprint.pprint(pd.merge(df6,df7,how='outer'))
#左結合
pprint.pprint("--------------------")
pprint.pprint(pd.merge(df6,df7,how='left'))
#右結合
pprint.pprint("--------------------")
pprint.pprint(pd.merge(df6,df7,how='right'))
#檢查缺失值
pprint.pprint("--------------------")
pprint.pprint(pd.merge(df6,df7,how='right').isnull().any())
pprint.pprint("--------------------")
temp_pd = pd.merge(df6,df7,how='right')
pprint.pprint(temp_pd[temp_pd['food'].isnull()])
#處理重複欄位名稱
df8 = pd.DataFrame({'name':['Bob','Jack','Lisa','Sue'],\
                    'rank':[1,2,3,4]})
df9 = pd.DataFrame({'name':['Bob','Jack','Lisa','Sue'],\
                    'rank':[3,1,4,2]})
pprint.pprint(df8)
pprint.pprint("--------------------")
pprint.pprint(df9)
pprint.pprint("--------------------")
pprint.pprint(pd.merge(df8,df9,on='name'))
#定義重複欄位名稱
pprint.pprint("--------------------")
pprint.pprint(pd.merge(df8,df9,on='name',suffixes=['_L','_R']))
```
